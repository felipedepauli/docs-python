# Começando baixo

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = np.arange(0, 1000)
y0 = np.sin(x0/50)
plt.plot(x0, y0)
```

```python
distance = np.random.randint(0, 20, 100)
print(distance)
frequency = np.random.randint(0, 100, 100)
print(frequency)
plt.scatter(distance, frequency)
```

Lindo! E se quisermos plotar algo em escala logarítmica?

```python
x1 = np.arange(0, 100)
y1 = x1 ** 2

plt.plot(x1, y1)
plt.show()
```

```python
y1 = x1 ** 2
plt.plot(x1, y1)
plt.xscale('log')
plt.show()
```

# Histograma

Histograma tem a ver com frequência. Conta-se quantas vezes cada valor aparece. O número contado é o domínio e a sua frequência é a imagem.

```python
x2 = np.random.randint(0, 100, 100)

print(x2.__len__())
print(len(x2))

plt.hist(x2)
```

```python
x2 = np.random.randint(0, 100, 50)

plt.hist(x2)
plt.show()


x2 = np.random.randint(0, 100, 50)

plt.hist(x2)
plt.clf()  # Limpa tudo que se escreveu na figura até agora.
plt.show()

x2 = np.random.randint(0, 100, 50)

plt.hist(x2, bins=5)
plt.show()
```

```python
x3 = np.arange(0, 20, 0.001)
y3 = 3 ** (-np.exp(x3/10)) * np.cos(x3)
plt.plot(x3, y3)

plt.xlabel('My pretty Axis X')
plt.ylabel('My sweet Axis Y')
plt.title('What a crazy function!')


plt.yticks( [-0.2, -0.05, 0, 0.05, 0.2],
            ["-0.2", "-1/20", "ZERO", "+1/20", "0.2"])

plt.ylim([-0.4, 0.4])

plt.text(1.5, 0.15, 'Such an amazing line!!!', color="orange")
plt.text(0.5, 0.3, 'Here the line begins', color="green")
plt.text(15, 0.02, 'Here de line ends', color="red")

plt.plot()
```
